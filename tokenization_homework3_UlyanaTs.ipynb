````python
import tiktoken
import matplotlib.pyplot as plt

enc = tiktoken.get_encoding('cl100k_base')

def stats(txt):
    words = txt.split()
    token_counts = [len(enc.encode(w)) for w in words]
    return {
        'total_words': len(words),
        'total_tokens': sum(token_counts),
        'avg_tokens_per_word': sum(token_counts)/len(words)
    }

eng_stats = stats(english_text)
ru_stats = stats(russian_text)
fr_stats = stats(french_text)
eng_stats, ru_stats, fr_stats
````

````python
def word_token_distribution(txt):
    words = txt.split()
    token_counts = [len(enc.encode(w)) for w in words]
    n = len(words)
    dist = {
        "1_token_pct": sum(1 for x in token_counts if x == 1) / n * 100,
        "2_tokens_pct": sum(1 for x in token_counts if x == 2) / n * 100,
        "3+_tokens_pct": sum(1 for x in token_counts if x >= 3) / n * 100
    }
    return dist

eng_dist = word_token_distribution(english_text)
ru_dist = word_token_distribution(russian_text)
fr_dist = word_token_distribution(french_text)

print("English word token distribution:", eng_dist)
print("Russian word token distribution:", ru_dist)
print("French word token distribution:", fr_dist)
````

````python
langs = ['English','Russian','French']
avgs = [eng_stats['avg_tokens_per_word'], ru_stats['avg_tokens_per_word'], fr_stats['avg_tokens_per_word']]
plt.bar(langs, avgs)
plt.title('Average Tokens per Word')
plt.show()
````

````python
def tokenize_text(txt):
    return enc.encode(txt)

english_tokens = tokenize_text(english_text)
russian_tokens = tokenize_text(russian_text)
french_tokens = tokenize_text(french_text)

print("English tokens:", english_tokens[:50], "... total:", len(english_tokens))
print("Russian tokens:", russian_tokens[:50], "... total:", len(russian_tokens))
print("French tokens:", french_tokens[:50], "... total:", len(french_tokens))
````
